```python
import pickle
import matplotlib.pyplot as plt
from pyscf import gto
import yaml
import numpy as np
import os
```

```python
with open('config_corrnet.yaml', 'r') as c:
    config = yaml.safe_load(c)
model_name = config['model_name']
```

```python
files = os.listdir("logging")
rel_files = [f for f in files if model_name in f]
rel_files.sort()
#directory = "logging/20230810_165359_densnet_Ael8lbyb_md_h2.npz_densities_final"  # Fill this in
# Get latest run
directory = f"logging/{rel_files[-1]}/preds.npz"
pred_data = np.load(directory, allow_pickle=True)
train_data = np.load("logging/20230816_184805_corrnet_FuBthqc9_md_h2.npz_corrs_final/best/best_loss.npz", allow_pickle=True)
mol_data = np.load("data/md_h2.npz", allow_pickle=True)
```

```python
mae = 0
for i in range(len(pred_data['data_idx'])):
    pred_corr = pred_data['pred_densities'][i]
    id = pred_data['data_idx'][i]
    real_corr = mol_data["corrs"][id]
    mae += np.abs((pred_corr - real_corr).sum()) / (166 ** 2)
mae = mae / (i + 1)
```
