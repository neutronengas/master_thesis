```python
import numpy as np
from data_utils import Y_m_l
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
r = np.random.normal(0, 1, (100000, 3))
r /= np.linalg.norm(r, axis=1)[:, None]
```

```python
m = -5.
l = 5.
lengths = np.abs(Y_m_l(r, m, l))
```

```python
[x, y, z] = (lengths[:, None] * r).T
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=lengths, marker='o')
plt.axis("equal")
plt.show()
```
